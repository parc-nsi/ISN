# Traitement d'une ligne

## Découpage d'une ligne selon le caractère de séparation  ';' 

```python
def decoupageLigne(ligne, sep):
    ligne = ligne.rstrip()
    return ligne.split(sep)
```

## Liste de liste de champs par année

```python
def listeVoeuxAnnee(fichier, annee, sep):
    voeux = []
    f = open(fichier)
    f.readline()  #sauter la ligne des en-tetes
    for ligne in f:
        champs = decoupageLigne(ligne, sep)
        if champs[0] == annee:
            voeux.append(champs)
    f.close()
    return voeux
```

```python
voeux_tous2016 = listeVoeuxAnnee(SOURCE, '2016', ';')
```

## Nombre total de candidats admis 

```python
def nombreAdmis(voeux):
    """Le nombe d'admis est dans le champ d'index 18"""
    s = 0
    for champs in voeux:
        val = champs[18]
        if val != 'inconnu':
            s = s + int(val)
    return s
    
```

```python
voeux_tous2017 = listeVoeuxAnnee(SOURCE, '2017',';')
len(voeux_tous2017)
```

```python
def nombreAdmisFilles(voeux):
    """Le nombe d'admis filles est dans le champ d'index 19"""
    s = 0
    for champs in voeux:
        val = champs[19]
        if val != 'inconnu':
            s = s + int(val)
    return s
```

# Filtrage des lignes par champs

## Filtrage des lignes selon une condition

On récupère toutes les lignes concernant l'établissement dont le  Code UAI (ou RNE) est passé en paramètre

```python
def filtreUai(voeux, codeUAI):
    voeux_filtre = []
    for v in voeux:        
        if v[1] ==  codeUAI:
            voeux_filtre.append(v)
    return voeux_filtre
```

## Filtrage générique sur  1 champ

```python
def filtre1champ(voeux, indexchamp, valchamp):
    t = []
    for v in voeux:        
        if v[indexchamp] ==  valchamp:
            t.append(v)
    return t
```

```python
filtre1champ(voeux_2016, 1, '0690026D') == filtreUai(voeux_2016, '0690026D')
```

## Filtrage générique sur  n champs

```python
def filtreNchamps(voeux, liste_indexchamp, liste_valchamp):
    t = []
    for v in voeux:   
        n = len(liste_indexchamp)
        match = True
        k = 0
        while k < n and match:
            indexchamp = liste_indexchamp[k]
            valchamp = liste_valchamp[k]
            if v[indexchamp] !=  valchamp:
                match = False
            else:
                k = k + 1
        if k == n:
            t.append(v)
    return t
```

```python
filtreNchamps(voeux_2016, [3,7], ['69','4_CPGE']) ==  filtreCpgeDepartement(voeux_2016, '69')
```

# Tri

## Classement des voeux CPGE 2016 dans l'ordre croissant du taux de boursiers

```python
voeux_cpge_2016 = filtreCpge(voeux_2016)

def clefTri(v):
    taux_boursier = v[32]
    if taux_boursier == '':
        taux_boursier = 0
    return float(taux_boursier)

tri_boursier_croissant_2016 = sorted(voeux_cpge_2016, key = clefTri)
```

```python
tri_boursier_croissant_2016[:10]
```

## Classement des voeux CPGE 2016 dans l'ordre décroissant du taux de boursiers

```python
tri_boursier_decroissant_2016 = sorted(voeux_cpge_2016, key = clefTri, reverse = True)

def affichage_tri(voeux):
    print('{:<5} {:<40} {:<10} {:<40} {:<20}'.format('Rang', 'Lycée', 'Département', 'Filière', 'Taux'))
    for k in range(len(voeux)):
        v = voeux[k]
        print('{:<5} {:<40} {:<10} {:<40} {:<20}'.format(k + 1, v[2][:40], v[3],v[9][:40],v[32]))
```

```python
affichage_tri(tri_boursier_decroissant_2016)
```

## Tri des voeux APB de l’année 2016 par ordre décroissant du taux de boursier admis puis par ordre alphabétique croissant de la filière (champ d’index 9)

## Fonction `filtreFiliere`

```python
def filtreFiliere(voeux, filiere):
    voeux_filtre = []
    for v in voeux:        
        if v[7] ==  filiere:
            voeux_filtre.append(v)
    return voeux_filtre
```

```python
tri_boursier_decroissant_MPSI = filtreFiliere(tri_boursier_decroissant_2016, 'MPSI')
```

## Séparation  des tris par taux décroissant d'admis boursiers par filière  puis écriture de chaque liste triée dans un fichier

```python
def tri_par_filiere(t):
    """Retourne une liste de liste, chaque sous-liste contient les voeux d'une filière de CPGE
    trié par taux d'admis boursiers décroissants.
    La liste t prise prise en argument est la liste de tous les voeux CPGE triés par 
    taux d'admis boursiers décroissants puis par filière."""
    filiere_courante = t[0][9]
    tri_filiere = [t[0]]
    tous_tris_filieres = [tri_filiere]
    for k in range(1, len(t)):
        filiere = t[k][9]
        if filiere == filiere_courante:
            tri_filiere.append(t[k])
        else:
            tous_tris_filieres.append(tri_filiere)
            filiere_courante = filiere
            tri_filiere = [t[k]]
    tous_tris_filieres.append(tri_filiere)
    return tous_tris_filieres
```

```python
def ecrire_fichier(voeux, fichier):
    f = open(fichier,'w')
    f.write('Lycée;Département;Filière;Taux de boursiers admis\n')
    for k in range(len(voeux)):          
        v = voeux[k]
        f.write(';'.join([v[2], v[3], v[9], v[32]]) + '\n')
    f.close()
```

```python
for tri_filiere in tri_par_filiere(tri_boursier_decroissant_2016_filiere):
    filiere = tri_filiere[0][9]
    ecrire_fichier(tri_filiere, 'tri_boursiers_'+filiere.replace("/","-")+'.csv')
```
